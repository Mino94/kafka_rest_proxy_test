```python
import requests
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

from urllib.parse import quote
from collections import Counter

import datetime
from datetime import timedelta
import random
import hashlib

from pymongo import MongoClient

HEADERS_LIST = """Mozilla/5.0 Slackware/13.37 (X11; U; Linux x86_64; en-US) AppleWebKit/534.16 (KHTML, like Gecko) Chrome/11.0.696.50
       Mozilla/5.0 Slackware/13.37 (X11; U; Linux x86_64; en-US) AppleWebKit/534.16 (KHTML, like Gecko) Chrome/12.0.742.91""".split(
    '\n')
HEADERS_LIST = [x.strip() for x in HEADERS_LIST]
agent = random.choice(HEADERS_LIST)
headers = {"User-Agent": "{}".format(agent)}

# 감정분석 api 변수 저장
emo_url = 'http://172.16.114.47:80/model-engine/social_senti_keyword/'


def doScrollDown(whileSeconds):
    start = datetime.datetime.now()
    end = start + datetime.timedelta(seconds=whileSeconds)

    while True:
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
        time.sleep(1)
        if datetime.datetime.now() > end:
            break


def delete_iframe(url):
    print(url)
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser')
    src_url = "https://blog.naver.com/" + soup.iframe["src"]
    print(src_url)
    return src_url


### headless로 실행하기 위한 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
options.add_argument("user-agent={}".format(agent))
options.add_argument("lang=ko_KR")

DRIVER_PATH = '/Users/kimminho/chromedriver'
driver = webdriver.Chrome(executable_path=DRIVER_PATH, chrome_options=options)


# mongodb connect
client = MongoClient('localhost', 27017)
db = client.naver
collection = db.blog_reply


keyword = "초록우산"
url = 'https://search.naver.com/search.naver?where=blog&query={}&sm=tab_opt&nso=so%3Ar%2Cp%3Afrom20210929to20210930'

driver.get(url)
driver.implicitly_wait(3)

doScrollDown(20)
html = driver.page_source

soup = BeautifulSoup(html, "html.parser")
posts = soup.find("ul", attrs={"class": "lst_total"}).find_all("li")

if not soup.find("div", attrs={"class": "api_noresult_wrap"}):
    posts = soup.find("ul", attrs={"class": "lst_total"}).find_all("li")
    for post in posts:

        comments_dict = {}
        try:
            post_title = post.find("a", attrs={"class": "api_txt_lines total_tit"}).get_text()
            post_link = post.find("a", attrs={"class": "api_txt_lines total_tit"})['href']
            if post_link:
                print(post_link)
                driver.get(delete_iframe(post_link))
                time.sleep(2)
                if driver.find_element_by_css_selector('#printPost1 > tbody > tr > td.bcc > div.post-btn.post_btn2 > div.wrap_postcomment > div.area_comment.pcol2').text == "댓글 쓰기":
                    print("댓글이 존재하지 않습니다.")
                    continue
                else:
                    driver.find_element_by_css_selector('#printPost1 > tbody > tr > td.bcc > div.post-btn.post_btn2 > div.wrap_postcomment > div.area_comment.pcol2').click()
                    time.sleep(3)
                    html2 = driver.page_source
                    soup2 = BeautifulSoup(html2, "html.parser")

                    comments_list = [p.text.replace('\n', '') for p in
                                     soup2.find_all("span", {"class": "u_cbox_contents"})]
                    nickname_list = [p.text for p in soup2.find_all("span", {"class": "u_cbox_nick"})]
                    dt_hour = [datetime.datetime.strptime(p.text, '%Y.%m.%d. %H:%M').strftime('%Y.%m.%d %H') for
                               p in soup2.find_all("span", {"class": "u_cbox_date"})]
                    date = [p['data-value'] for p in soup2.find_all("span", {"class": "u_cbox_date"})]

                    for i, comment in enumerate(comments_list):

                        temp_id = str(post_title + str(date) + nickname_list[i])
                        temp_id = temp_id.encode('utf-8')
                        h = hashlib.new('md5')
                        h.update(temp_id)
                        _id = h.hexdigest()

                        # 감정분석 추가
                        try:
                            emo_response = requests.post(emo_url, json={'text': [comment]}, timeout=5)
                            emotions = emo_response.json()
                            source_emo = emotions['pred_labels'][0]
                            key_rk = emotions['keyword'].get("Noun_keywords")[0]
                            keyword_rk = dict(Counter(key_rk))
                            key_ak = emotions['keyword'].get("Not_Noun_keywords")[0]
                            keyword_ak = dict(Counter(key_ak))
                        except:
                            source_emo = None
                            keyword_rk = {}
                            keyword_ak = {}

                        comments_dict = {
                            '_id': _id,
                            'title': post_title,
                            'keyword': keyword,
                            'date': date[i],
                            'dt_hour': dt_hour[i],
                            'body': comment,
                            'user_name': nickname_list[i],
                            'portal': "naver",
                            "count_ak": keyword_ak,
                            "count_rk": keyword_rk,
                            "emotions": source_emo,
                            'url': post_link
                        }
                        print(comments_dict)
                        try:
                            collection.insert_one(comments_dict)
                            print("insert 완료!!!!!!!!!!!!!!!!")
                        except Exception as e:
                            print("error : ", e)
                            continue
            else:
                print("no postlinke")

        except Exception as e:
            print(e)
            continue
else:
    print("블로그 게시글이 없습니다.")

driver.close()
```
